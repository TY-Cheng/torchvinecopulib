```python
# add autoreload and other ipython magic
import pandas as pd
from IPython import get_ipython
from tqdm import tqdm
from vcae.config import config
from vcae.experiment import run_experiment

ipython = get_ipython()
ipython.run_line_magic("load_ext", "autoreload")
ipython.run_line_magic("autoreload", "2")
ipython.run_line_magic("matplotlib", "inline")
ipython.run_line_magic("config", 'InlineBackend.figure_format = "retina"')

results = []
# dataset = "MNIST"  # or "SVHN"
dataset = "SVHN"  # or "MNIST"
use_mmd = False  # or True

for seed in tqdm(range(1), desc="Running experiments"):
    result = run_experiment(seed, config, dataset=dataset, use_mmd=use_mmd)
    results.append(result)


df_results = pd.DataFrame(results)
# df_results.to_csv("experiment_results.csv", index=False)
```
